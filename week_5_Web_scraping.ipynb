```python
import pandas as pd
```

```python
# To get the site's HTML code into your python script to that you can interact with it. For this task, we will use python's request library.
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url= 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
```

## steps in webscraping
## 1. Inspect your data source
## 2.scrap the HTML content from a page
## 3. Parse the Html code

## Send the HTTP request to the URL of the webpage  you want to access.

```python
# This code issues an HTTP GET request to the given URL.
# It retrieves the HTML data that the server sends back and stores that data in a python object.
response=requests.get(url)
```

```python
response # if it is 200 means it is correct
```

## Parsing the data

```python
soup= BeautifulSoup(response.content,'html.parser')
```

## create empty list to store data

```python
movie_name=[]
year=[]
time=[]
rating=[]
```

## searching for the data


```python
movie_data=soup.findAll("div",attrs={"class":"lister-item mode-advanced"})
```

```python
movie_data
```

## Extracting data


```python
for store in movie_data:
  name=store.h3.a.text
  movie_name.append(name)
  yr=store.h3.find("span",class_="lister-item-year text-muted unbold").text.replace("(","").replace(")","")
  year.append(yr)
  mov_min=store.p.find("span",class_="runtime").text.replace("min","").replace(" ","")
  time.append(mov_min)
  rat=store.find("div",class_="inline-block ratings-imdb-rating").text.replace("\n","")
  rating.append(rat)


```

```python
movie_name
```

```python
year
```

```python
time
```

```python
rating
```

## converting data into Dataframe

```python
movie_dataset=pd.DataFrame({"movie_name":movie_name,"year":year,"movie_rating":rating,"time":time})
```

```python
movie_dataset
```

## save the data frame

```python
movie_dataset.to_csv("movie.csv")
```
